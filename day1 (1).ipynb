```python
prompt_sort = '''I will pass m numbers of natural numbers from 1 to n as a list. Please proceed in the order below.
1. Select the fastest list sorting algorithm
2. Write it as Python code and actually run it.
3. Print the results.

[Example]
Input: [5, 4, 2, 1, 6]
Output: [1, 2, 4, 5, 6]

Input: '''
```

```python
prompt_sort = '''- You are the most accurate algorithmic function.
- Sort the given number array in ascending order and show it again.
- Sort using Quick Sort.
- Show only the results without adding additional explanation.

[Example 1]
Input: [4, 1, 31, 9]
Output: [1, 4, 9, 31]

[Example 2]
Input: [20, 33, 40, 9]
Output: [9, 20, 33, 40]

[Example 3]
Input: [39, 12, 31, 29]
Output: [2, 29, 31, 39]

[Example 4]
Input: [55, 22, 1, 99]
Output: [1, 22, 55, 99]


Input: '''
```

```python
completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": prompt_sort},
    {"role": "user", "content": "[10, 20, 33, 7, 21]"}
  ]
)

print(completion.choices[0].message)
```

```python
from llmlingua import PromptCompressor

llm_lingua = PromptCompressor(
    model_name="microsoft/llmlingua-2-xlm-roberta-large-meetingbank",
    use_llmlingua2=True,
)
```

```python
# 2000 Compression
compressed_prompt = llm_lingua.compress_prompt(
    prompt_sort,
    rate=0.33,
    force_tokens=["!", ".", "?", "\n"],
    drop_consecutive=True,
)
```

```python
import json

question = "[10, 20, 33, 7, 21]"

prompt = "\n\n".join([compressed_prompt["compressed_prompt"], question])

message = [
    {"role": "user", "content": prompt}
]


completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages= message
)


print(completion.choices[0].message)
#print(json.dumps(response, indent=4))
```
